```python
from PIL import Image

img = Image.open('oxygen.png')

width = img.width
height = img.height

y = height // 2

values = []

for x in range(0, width):
    test = img.getpixel((x,y))
    print(test[0])
    values.append(test[0])

print(values)
```

```python
new = ''

for count, i in enumerate(values):
    if count == 0:
        new += chr(i)
    elif chr(i).isnumeric():
        new += chr(i)
    elif values[count] != values[count - 1]:
        new += chr(i)
    

print(new)
```

```python
next_page = [105, 110, 116, 101, 103, 114, 105, 116, 121]
word = [chr(x) for x in next_page]
new = ''
for i in word:
    new += i

print(new)
```
