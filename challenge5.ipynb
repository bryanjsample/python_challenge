```python
import requests

from bs4 import BeautifulSoup

url = 'http://www.pythonchallenge.com/pc/def/peak.html'

req = requests.get(url)
soup = BeautifulSoup(req.content, 'html.parser')

print(soup.prettify())
```

```python
import urllib

new_url = 'http://www.pythonchallenge.com/pc/def/banner.p'

mess = urllib.request.urlopen(new_url).read()
text = mess.decode()

print(text)
```

```python
import pickle

data = pickle.load(urllib.request.urlopen(new_url))

print(data)

for i in data:
    print(''.join([k * v for k , v in i]))
```
