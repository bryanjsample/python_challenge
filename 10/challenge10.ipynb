```python
a = ['1']

# one one, two one, one two one one, three one two one one one

def num_count(num):
    while len(a) < 31:
        same_nums = ''
        nums_lyst = []
        for index, i in enumerate(num):
            if index == 0:
                same_nums += i
                if index == len(num) - 1:
                    nums_lyst.append(same_nums)
            else:
                if i == num[index - 1]:
                    same_nums += i
                    if index == len(num) - 1:
                        nums_lyst.append(same_nums)
                else:
                    nums_lyst.append(same_nums)
                    same_nums = '' + i
                    if index == len(num) - 1:
                        nums_lyst.append(same_nums)
            good_string = ''
            for x in nums_lyst:
                good_string += str(len(x)) + x[0]

        a.append(good_string)
        print(a)

        num_count(good_string)

num_count('1')

print(len(a[30]))



```
