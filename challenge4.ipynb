```python
import requests

from bs4 import BeautifulSoup

url = 'http://www.pythonchallenge.com/pc/def/linkedlist.php'

req = requests.get(url)
soup = BeautifulSoup(req.content, 'html.parser')

print(soup.prettify())

first_tag = soup.find('a').get('href')

stripped_url = url.replace('linkedlist.php', '')

new_page = stripped_url + first_tag

print(new_page)
```

```python
import urllib.request

for i in range(0, 400):
    req = urllib.request.urlopen(new_page)
    text = str((req.read()))
    text_list = text.split('is')
    target = text_list[-1].replace(' ', '').replace("'", '')
    print(target)
    new_page = stripped_url + 'linkedlist.php?nothing=' + target
```
