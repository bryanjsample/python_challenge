```python
import os

parent_dir = os.getcwd()

dir = parent_dir + '/channel_c6'

file = '90052.txt'

for x in range(0,910):
    with open(f'{dir}/{file}') as f:
        contents = f.read()
        print(contents, file, sep = ' | ')
        contents_list = contents.split()
    for i in contents_list:
        if i.isnumeric():
            file = f'{i}.txt'
```

```python
import zipfile

zipped = zipfile.ZipFile(dir + '.zip', 'r')

zip_i = zipped.namelist()

message = []
for i in zip_i:
    x = zipped.getinfo(i).comment.decode()
    message.append(x)

for i in message:
    print(i)
```

```python
letters = {}
for char in message:
    if letters.get(char, 0) == 0:
        letters[char] = 1
    else:
        letters[char] += 1

print(letters)
```
