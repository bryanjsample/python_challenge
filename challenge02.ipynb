```python
import requests

req = requests.get('http://www.pythonchallenge.com/pc/def/ocr.html')

mess = req.text

print(mess)
```

```python
pieces = mess.split('<!--')
target = pieces[-1].replace('-->', '')
occurs = {}

for i in target:
    if occurs.get(i, 0) == 0:
        occurs[i] = 1
    else:
        occurs[i] += 1

print(occurs)
```

```python
for letter, num in occurs.items():
    if num != 1:
        target = target.replace(letter, '')

print(target)
```
